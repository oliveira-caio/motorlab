```python
import os

from pathlib import Path

import matplotlib.pyplot as plt
import motorlab as ml
import numpy as np
import yaml
```

```python
os.chdir(Path().resolve().parent)
```

```python
POSES_DIR = "data/gbyk/bex_20230621_spikes_sorted_SES/poses"
poses = ml.data.load_from_memmap(POSES_DIR).reshape(-1, 21, 3)
```

```python
d = np.zeros(poses.shape[0])

for s, e in ml.poses.skeleton:
    s_idx = ml.poses.keypoints["gbyk"][s]
    e_idx = ml.poses.keypoints["gbyk"][e]
    d += np.linalg.norm(poses[:, s_idx] - poses[:, e_idx], axis=1)
```

```python
plt.hist(d, 10)
plt.show()
```
